#mWDN Traffic

```python
with open("/content/drive/MyDrive/model_traffic.pt", 'rb') as f:
    model = torch.load(f, map_location=torch.device('cpu'))

dSet = Datasets.Traffic().data
d = utils.Data_utility(dSet, 0.6, 0.2, False, 12, 168, 2);
d.train[0]=d.train[0].permute(0,2,1)
d.valid[0]=d.valid[0].permute(0,2,1)
d.test[0]=d.test[0].permute(0,2,1)

model_name = "mWDN"
name = "Traffic"
```

#SHAP Values

```python
explainer = shap.DeepExplainer(model, d.valid[0][0:10])
shap_values = explainer.shap_values(d.valid[0][0:10])
```

```python
shap_total = shap_values[0].copy()
for s in shap_values:
    shap_total = shap_total + np.abs(s)
shap_total = shap_total - shap_values[0]
ax = sns.heatmap(shap_total[0])
```

# Top 5 Most Influential Features

```python
# ind[5] contains the indices of the Top 5 most influential time series (in order)
ind = np.sum(shap_total[0],1).argsort()[-5:][::-1]

colormap = cm.Reds
normalize = mpl.colors.Normalize(vmin=shap_total[0].min(), vmax=shap_total[0].max())
fig, ax = plt.subplots(5, figsize=(12, 3), dpi=300, facecolor='w', edgecolor='k')

for j in range(len(ind)):
    i = ind[j]
    ax[j].plot(range(1,169), d.valid[0][0][i], color = "blue", label=str(i+1))
    ax[j].set_axis_off()
    ax[j].legend(loc="upper right")
    for t in range(168):
      ax[j].fill_between([t+0.5,t+1.5], [d.valid[0][0][i].max(),d.valid[0][0][i].max()], [d.valid[0][0][i].min(),d.valid[0][0][i].min()], color=colormap(normalize(shap_total[0][i][t])))
      ax[j].set_ylim([d.valid[0][0][i].min()-d.valid[0][0][i].numpy().ptp()*0.015,d.valid[0][0][i].max()+d.valid[0][0][i].numpy().ptp()*0.015])

plt.subplots_adjust(hspace=0)
fig.show()
plt.savefig("stats/"+model_name+"_"+name+".png", bbox_inches='tight')
```

#Perturbation Sensitivity

```python
perturbed_input = d.valid[0][0:10].clone()
for example in range(10):
    threshold = np.percentile(shap_total[example], 90)
    for i in range(shap_total.shape[1]):
        for j in range(shap_total.shape[2]):
            if shap_total[example][i][j] >= threshold:
                r1 = max(j-12,0)
                r2 = min(j+13,shap_total.shape[2])
                perturbed_input[example][i][j] = torch.mean(d.valid[0][example][i][r1:r2])
errors = torch.norm(model(d.valid[0][0:10])-model(perturbed_input), dim=1)
f = open("stats/"+model_name+"_"+name+".txt","w") 
for e in errors:
    f.write(str(e.item()) + '\n')
f.close()
```
