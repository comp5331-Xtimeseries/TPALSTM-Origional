#SHAP Values

```python
plt.figure(num=None, figsize=(12, 3), dpi=300, facecolor='w', edgecolor='k')
sns.heatmap(np.sum(shap_total,2), cmap="YlGnBu")
plt.title(name + " Dataset")
plt.xlabel("Feature")
plt.ylabel("Example")
plt.savefig("stats/"+model_name+"_"+name+"_heat.png", bbox_inches='tight')
```

# Top 3 Most Influential Features

```python
# ind[3] contains the indices of the Top 3 most influential time series (in order)
ind = np.sum(shap_total[0],1).argsort()[-3:][::-1]

colormap = cm.Reds
normalize = cm.colors.Normalize(vmin=shap_total[0].min(), vmax=shap_total[0].max())
fig, ax = plt.subplots(3, figsize=(12, 3), dpi=300, facecolor='w', edgecolor='k')

for j in range(len(ind)):
    i = ind[j]
    ax[j].plot(range(1,169), d.valid[0][0][i], color = "blue", label=str(i))
    ax[j].xaxis.set_ticks([])
    ax[j].yaxis.set_ticks([])
    ax[j].legend(loc="upper right")
    for t in range(168):
      ax[j].fill_between([t+0.5,t+1.5], [d.valid[0][0][i].max(),d.valid[0][0][i].max()], [d.valid[0][0][i].min(),d.valid[0][0][i].min()], color=colormap(normalize(shap_total[0][i][t])))
      ax[j].set_ylim([d.valid[0][0][i].min()-d.valid[0][0][i].numpy().ptp()*0.015,d.valid[0][0][i].max()+d.valid[0][0][i].numpy().ptp()*0.015])

fig.text(0.5, 0.90, name+' Dataset', ha='center', fontsize=13)
fig.text(0.5, 0.05, 'Time', ha='center', fontsize=11)
fig.text(0.10, 0.5, 'Top 3 Features', va='center', rotation='vertical', fontsize=11)
plt.subplots_adjust(hspace=0)
fig.show()
plt.savefig("stats/"+model_name+"_"+name+"_top3.png", bbox_inches='tight')
```

#Perturbation Sensitivity

```python
perturbed_input = d.valid[0][0:10].clone()
for example in range(10):
    threshold = np.percentile(shap_total[example], 90)
    for i in range(shap_total.shape[1]):
        for j in range(shap_total.shape[2]):
            if shap_total[example][i][j] >= threshold:
                r1 = max(j-24,0)
                r2 = min(j+25,shap_total.shape[2])
                perturbed_input[example][i][j] = torch.mean(d.valid[0][example][i][r1:r2])
errorsOriginal = torch.norm(d.valid[1][0:10]-model(d.valid[0][0:10]), dim=1)
errorsNoisy = torch.norm(d.valid[1][0:10]-model(perturbed_input), dim=1)
f = open("stats/"+model_name+"_"+name+".txt","w") 
f.write("errorsOriginal" + '\t' + "errorsNoisy" + '\t' + "Percentage Difference" + "\n")
print("errorsOriginal" + '\t' + "errorsNoisy" + '\t' + "Percentage Difference")
s = 0
for i in range(10):
    f.write(str(errorsOriginal[i].item()) + '\t' + str(errorsNoisy[i].item()) + '\t' + str(100*(errorsNoisy[i].item()/errorsOriginal[i].item()-1)) + '\n')
    print(str(errorsOriginal[i].item()) + '\t' + str(errorsNoisy[i].item()) + '\t' + str(100*(errorsNoisy[i].item()/errorsOriginal[i].item()-1)))
    s += 100*(errorsNoisy[i].item()/errorsOriginal[i].item()-1)
f.write(str(s/10))
f.close()
print(s/10)
```
